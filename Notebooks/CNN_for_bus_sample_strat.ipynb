```python
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/bus_sample 2.csv")
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.2)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense

X_train = train.drop(['TimeToArrival','ExpectedHour','ExpectedDayOfWeek'], axis=1).values
y_train = train['TimeToArrival'].values
X_test = test.drop(['TimeToArrival','ExpectedHour','ExpectedDayOfWeek'], axis=1).values
y_test = test['TimeToArrival'].values

# Reshape data for CNN (assuming each sample is a sequence of features)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

# Define the CNN model
model = Sequential()
model.add(Conv1D(32, 3, activation='relu', input_shape=(X_train.shape[1], 1)))
model.add(MaxPooling1D(2))
model.add(Conv1D(64, 2, activation='relu',padding='same')) # Add padding to the second convolutional layer
#Removed the second MaxPooling1D layer to avoid negative dimension size
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32)

# Evaluate the model
loss = model.evaluate(X_test, y_test)
print("Mean Squared Error on Test Set:", loss)

# Make predictions
predictions = model.predict(X_test)
```

```python
for i in range(len(predictions)):
  print("Predicted:", predictions[i][0], "Actual:", y_test[i])
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

y_pred = model.predict(X_test)

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

# Print the results
print('Mean Absolute Error:', mae)
print('Mean Squared Error:', mse)
print('Root Mean Squared Error:', rmse)
print('R-squared:', r2)
```

```python
model.save('cnn_model.h5')
```
